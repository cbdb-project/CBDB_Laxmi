```python
import pandas as pd
from tqdm import tqdm
import warnings
warnings.filterwarnings("ignore")
```

```python
office_title_list=pd.read_excel('method1.xlsx', sheetname='raw_verified')['官'].dropna().reset_index().drop('index', axis=1)
# Drop empty values.
office_title_collection=[]
for value in office_title_list['官']:
    office_title_collection+=value.split('；')
office_title_collection=[s for s in office_title_collection if s!='']

to_do_file=pd.read_excel('method2.xlsx', sheetname='raw_indexed', index_col=0)
```

```python
to_do_file
```

```python
import ipyparallel as ipp
c = ipp.Client()
print (c.ids)
dview = c[:]

dview.execute('import pandas as pd')
dview['to_do_file']=to_do_file
dview['office_title_collection']=office_title_collection
dview['df_tagged']=pd.DataFrame()
```

```python
@dview.parallel(block=True)
def tag_office_title(index):
    title_matched=[]
    global df_tagged
    for office_title in office_title_collection:
        if office_title in str(to_do_file.loc[index, '信息']) and office_title not in title_matched:
            title_matched.append(office_title)
    if title_matched!=[]:
        for title in title_matched:
            df_temp=to_do_file.loc[index]
            df_temp['tag_office_title']=title
            df_tagged=df_tagged.append(df_temp)
    else:
        pass
```

```python
t=tag_office_title.map(to_do_file.index)
```

```python
pd.concat(dview.gather('df_tagged')).sort_index().to_excel('method2_tagged.xlsx')
```
